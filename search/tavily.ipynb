## Tavily를 사용한 비동기 API 호출

```python
import asyncio
from tavily import AsyncTavilyClient

# Initialize Tavily client
tavily_client = AsyncTavilyClient(os.getenv("TAVILY_API_KEY"))

async def fetch_and_gather():
    queries = ["latest AI trends", "future of quantum computing"]

    # Perform search and continue even if one query fails (using return_exceptions=True)
    try:
        responses = await asyncio.gather(*(tavily_client.search(q) for q in queries), return_exceptions=True)

        # Handle responses and print
        for response in responses:
            if isinstance(response, Exception):
                print(f"Search query failed: {response}")
            else:
                print(response)

    except Exception as e:
        print(f"Error during search queries: {e}")
```

```python
# jupyter notebook에서 코드를 실행했을 때 안되고, 파이썬 파일로 코드를 실행하니 정상적으로 작동
# "예기치 않은 동작으로 이어질 수 있는 중첩된 이벤트 루프를 실수로 생성하는 것을 방지하기 위해" asyncio가 적용해 놓은 실행환경 제한 에러
# Run the function
asyncio.run(fetch_and_gather())
```

```python
# 셀 3: TavilyExtract 초기화 및 호출
from langchain_tavily import TavilyExtract

# Extract API 도구 설정
tool = TavilyExtract(
    extract_depth="basic",   # "basic" 또는 "advanced"
    include_images=True      # 페이지 내 모든 이미지 URL 포함
)

# 분석할 URL 리스트
urls = ["https://naver.com"]

# API 호출
result = tool.invoke({"urls": urls})
```

```python
print(result)
```

```python
for site in result["results"]:
    print(site["images"])

```

```python
# 셀 4: 파비콘(favicon) URL만 필터링 출력
# result 구조에 따라 조정할 수 있습니다.
for site in result["results"]:
    images = site.get("images", [])
    # 흔히 파비콘은 파일명에 'favicon'이 들어갑니다.
    favicons = [img for img in images if "favicon" in img.lower()]
    print(f"사이트: {site.get('url')}")
    if favicons:
        for ico in favicons:
            print("  • Favicon URL:", ico)
    else:
        print("  • Favicon 정보를 찾을 수 없습니다.")
```

```python
# (선택) 셀 5: PIL을 이용해 직접 이미지 확인하기
from PIL import Image
import requests
from io import BytesIO

if favicons:
    response = requests.get(favicons[0])
    img = Image.open(BytesIO(response.content))
    display(img)  # 노트북에서 이미지 표시
```

```python
# 셀 2: 라이브러리 임포트
from tavily import TavilyClient
from bs4 import BeautifulSoup
from urllib.parse import urljoin
import os
from dotenv import load_dotenv

load_dotenv()
```

```python
# 셀 3: Search 호출
client = TavilyClient(api_key=os.getenv("TAVILY_API_KEY"))
resp = client.search("site:naver.com", include_raw_content=True)
```

```python
print(resp)
```

```python
# 셀 4: HTML 파싱 함수
def parse_favicons(html, base_url):
    soup = BeautifulSoup(html, "html.parser")
    links = soup.find_all("link", rel=lambda v: v and "icon" in v.lower())
    return [urljoin(base_url, l["href"]) for l in links if l.get("href")]
```

```python
# 셀 5: 각 결과에서 파비콘 추출
for item in resp["data"]:
    html = item.get("raw_content", "")
    favs = parse_favicons(html, item["url"])
    print(item["url"], "→", favs or ["없음"])
```
